```python
def imshow(img, size=8):
    import math
    
    if type(img) is not np.ndarray:
        img = img.detach().cpu().numpy()

    num_dims = len(img.shape)

    if num_dims == 4:
        bs = img.shape[0]
        channels = img.shape[1]

        rows = math.ceil(math.sqrt(bs))
        cols = math.ceil(bs / rows)
        new_bs = rows * cols
        H, W = img.shape[2], img.shape[3]

        img = np.pad(img, ((0, new_bs - bs), (0, 0), (0, 0), (0, 0)))
        img = img.reshape(rows, cols, channels, H, W)
        img = np.concatenate(np.split(img, rows, axis=0), axis=3)
        img = np.concatenate(np.split(img, cols, axis=1), axis=4)
        img = img[0, 0]

    img = np.transpose(img, (1, 2, 0))
    if img.shape[-1] == 1:
        img = img[:, :, 0]
    
    fig, ax = plt.subplots(figsize=(size, size))
    ax.imshow(img)
    plt.tight_layout()
```

```python
class GlyphEncoder(nn.Module):
    def __init__(self, msg_size, img_channels):
        super().__init__()

        self.net = nn.Sequential(
#             nn.Embedding(num_glyphs, 32),
            tu.reshape(-1, msg_size, 1, 1),
            tu.deconv_block(msg_size , 128         , ks=5, s=2, p=1),
            nn.Dropout(0.2),
            tu.deconv_block(128, 64          , ks=5, s=1, p=2),
            nn.Dropout(0.5),
            tu.deconv_block(64 , 32          , ks=5, s=1, p=2),
            tu.deconv_block(32 , 16          , ks=5, s=2, p=1),
            nn.Dropout(0.3),
            tu.deconv_block(16 , 8           , ks=5, s=1, p=2),
            tu.deconv_block(8  , 8           , ks=5, s=1, p=2),
            nn.Dropout(0.01),
            tu.deconv_block(8  , 4           , ks=5, s=2, p=1),
            tu.deconv_block(4  , img_channels, ks=4, s=2, p=0, a=nn.Sigmoid()),
        )

    def forward(self, z):
        return self.net(z)
    

msg_size = 32
glyph_encoder = GlyphEncoder(msg_size, img_channels=3)
msg = T.rand(20, msg_size)
fake_img = glyph_encoder(msg)

imshow(fake_img, size=6)
fake_img.shape
```

```python
class GlyphDecoder(nn.Module):
    def __init__(self, in_channels, msg_size):
        super().__init__()
        
        self.net = nn.Sequential(
            tu.conv_to_flat(
                [in_channels, 128, 128, 64, 32, 16],
                msg_size,
                ks=3,
                s=2,
            ),
        )
        
    def forward(self, x):
        return self.net(x)
    
bs = 16
in_channels = 3
msg_size = 32

glyph_decoder = GlyphDecoder(in_channels, msg_size)
random_glyphs = T.rand(bs, in_channels, 32, 32)
decoded_glyphs = glyph_decoder(random_glyphs)
decoded_glyphs.shape
```

```python
class GlyphAE(tu.BaseModule):
    def __init__(self, msg_size, img_channels):
        super().__init__()
        bs = 3
        self.msg_size = msg_size
        self.encoder = GlyphEncoder(msg_size, img_channels)
        self.decoder = GlyphDecoder(img_channels, msg_size)

        # This is done so that te decoder parameters are initialized
        imgs = self.encoder(self.sample(bs=1))
        self.decoder(imgs)

    @property
    def device(self):
        return next(self.encoder.parameters()).device

    def sample(self, bs):
        return T.randn(bs, self.msg_size).to(self.device)

    def forward(self, bs):
        msg = self.sample(bs)
        img = self.encoder(msg)
        return img

    def configure_optim(self, lr, bs, noise_size):
        self.noise_size = noise_size
        self.bs = bs
        self.optim = T.optim.Adam(self.parameters(), lr)

    def apply_noise(self, t):
        noise = T.normal(0, 1, t.shape)
        noise = noise.to(self.device)
        return t * noise

    def optim_step(self, *_):
        self.optim.zero_grad()
        msg = self.sample(self.bs)
        
#         with ef.scan():
        img = self.encoder(msg)
        img = self.apply_noise(img)
        pred_msg = self.decoder(img)

        loss = F.mse_loss(pred_msg, msg)
        loss.backward()

        self.optim.step()

        return loss.item(), {
            'msg': msg,
            'pred_msg': pred_msg,
            'img': img
        }
```

```python
bs = 16
num_glyphs = 15
glyph_ae = GlyphAE(num_glyphs, img_channels=3)

for _ in range(5):
    glyph_ae.configure_optim(lr=0.001, bs=128, noise_size=0.5)
    loss, info = glyph_ae.optim_step()

fake_glyphs = glyph_ae(16)
imshow(fake_glyphs, size=5)
```

```python
num_glyphs = 50
glyph_ae = GlyphAE(num_glyphs, img_channels=1)
glyph_ae = glyph_ae.to('cuda')
```

```python
its = 512
epochs = 5

for e in range(epochs):
    tr = trange(its)
    for _ in tr:
        glyph_ae.configure_optim(lr=0.0002, bs=128, noise_size=1 / (epochs - e))
        loss, info = glyph_ae.optim_step()
        tr.set_description(f'Loss: {loss:0.6f}')

    fake_glyphs = glyph_ae(9)
    imshow(fake_glyphs, size=8)
    plt.show()
```
